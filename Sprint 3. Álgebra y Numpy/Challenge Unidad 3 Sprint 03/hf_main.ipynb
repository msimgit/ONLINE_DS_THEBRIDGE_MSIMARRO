```python
# Importamos todas las librerías que necesitaremos
import numpy as np
import random
import hf_clas as clas
import hf_var as var
import hf_fct as fct

def iniciar_juego():
    print(var.mensajeinicial)
    nombre = input("Introduce tu nombre, jugador: ")
    print(f"\n¡Bienvenido, {nombre}!\n")

    tablero_jp = clas.Tablero()
    tablero_cpu = clas.Tablero()

    lado = 10
    tablero_jp.crear_tablero(lado)
    tablero_cpu.crear_tablero(lado)
    tablero_visible = np.full((lado, lado), " ")

    flota = { "Destructor": 4, "Submarino": 3, "Acorazado": 2, "Portaaviones": 1 }
    tablero_jp.colocar_barcos(flota)
    tablero_cpu.colocar_barcos(flota)

    print("Tu tablero:")
    tablero_jp.mostrar()
    print("El tablero de tu contrincante: (se muestra vacío)")
    
    ronda = 0

    while True:
        ronda += 1
        print(f"\n--- Ronda {ronda} ---")
        turno_jugador = True
        while turno_jugador == True:
            
            print("\nTu tablero:")
            tablero_jp.mostrar()
            fct.mostrar_tablero_cpu_visible(tablero_cpu, tablero_visible)

            print(f"\n{var.mensajeturno}")
            entrada_fila = input("Introduce la fila (0-9) o 'S' para salir: ").strip().upper()
            if entrada_fila == 'S':
                print(var.mensajesalida)
                return

            try:
                x = int(entrada_fila)
                y = int(input("Introduce la columna (0-9): "))
            except ValueError:
                print("Entrada no válida. Intenta de nuevo.")
                continue

            resultado = tablero_cpu.disparar(x, y)

            if resultado == "impacto": # El jugador sigue jugando
                print(var.disparoacertado)
                if tablero_cpu.comprobar_victoria():
                    print(var.mensajevictoria)
                    return
                continue
                
            elif resultado == "agua":
                print(var.disparofallido)
                break
                ronda += 1
            elif resultado == "repetido":
                print("Ya has disparado ahí. Pierdes turno.")
                break
        
        turno_cpu = True
        while turno_cpu == True:
            print("\nTurno de la CPU:")
           
            x = random.randint(0, 9)
            y = random.randint(0, 9)
            resultado = tablero_jp.disparar(x, y)

            if resultado == "repetido":
                print("La CPU ha repetido una tirada. Vuelve a intentarlo.")
                continue
            elif resultado == "impacto": # La CPU sigue disparando
                print("¡La CPU ha acertado!")
                if tablero_jp.comprobar_victoria():
                    return var.mensajederrota
                continue
            elif resultado == "agua":
                print("La CPU ha fallado.")
                turno_cpu = False

    print(var.mensajefinal)

if __name__ == "__main__":
    iniciar_juego()
```
